```python
with open ("C:/Users/vedha/Downloads/t20_wc_batting_summary.json") as f:
  data=json.load(f)
    
  all_records=[]

  for rec in data:
     all_records.extend(rec["battingSummary"])

     
df_batting=pd.DataFrame(all_records)
print(df_batting.head(11))
print(df_batting.tail())
print (df_batting.shape)
```

```python
#creating a not out /out column from dismissal column and then dropping dismissal column

df_batting["out/not_out"]=df_batting.dismissal.apply(lambda x:"out" if len(x) >0 else "not_out")

df_batting.drop(columns=["dismissal"],inplace=True)
df_batting.head(11)
```

```python
# Removing special characters like the one after Mendis

df_batting["batsmanName"]=df_batting.batsmanName.apply(lambda x: x.replace("â€"," "))
df_batting["batsmanName"]=df_batting.batsmanName.apply(lambda x: x.replace("\xa0"," "))

df_batting.head(11)
```

```python
# Creating a match id dictionary for a key column because in Batting summary and match summary data we dont have any common column

match_id_dict={}

for index, row in df_match.iterrows():
    key1 = row['team1'] +" Vs " + row['team2']
    key2 = row['team2'] +" Vs " + row['team1']

    match_id_dict[key1]= row["match_id"]
    match_id_dict[key2]= row["match_id"]

match_id_dict
```

```python
#Creating a new column of match id in batting summary data
df_batting["match_id"]=df_batting["match"].map(match_id_dict)
df_batting.head(11)
```
